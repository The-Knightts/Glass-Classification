```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score         
from sklearn.model_selection import train_test_split
```

```python
x = dataset.drop(columns = ['Type'])
y = dataset['Type']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state = 1)
```

```python
clf = RandomForestClassifier(n_estimators = 100, random_state = 42)
```

```python
print(x)
```

```python
print(y)
```

```python
clf.fit(xtrain,ytrain)
```

```python
xtrain.count()
```

```python
ytrain.count()
```

```python
ypred = clf.predict(xtest)
```

```python
ypred
```

```python
accuracy = accuracy_score(ytest, ypred)
print(f'Accuracy: {accuracy:.2f}')
```

```python
def predict_glass_type(clf, new_data):
   
    new_df = pd.DataFrame(new_data)

    predicted_labels = clf.predict(new_df)
    
    glass_names = {
        1: "building_windows_float_processed",
        2: "building_windows_non_float_processed",
        3: "vehicle_windows_float_processed",
        4: "vehicle_windows_non_float_processed",
        5: "containers",
        6: "tableware",
        7: "headlamps"
    }
    
    predicted_label = predicted_labels[0]
    predicted_name = glass_names.get(predicted_label)
    
    return predicted_label, predicted_name

new_data = {
    'RI': [1.52],
    'Na': [13.64],
    'Mg': [4.49],
    'Al': [1.10],
    'Si': [71.78],
    'K':  [0.06],
    'Ca': [8.75],
    'Ba': [0.0],
    'Fe': [0.0]
}

predicted_label, predicted_name = predict_glass_type(clf, new_data)
print(f'Predicted Glass Type (Label): {predicted_label}')
print(f'Predicted Glass Type (Name): {predicted_name}')
```
